### Exercise 8 - Distribution of scores

The restaurant letter grades are based off of their scores in the following way:
- A score of less than 14 points results in an A grade.
- A score of 14-27 points means a restaurant receives a B grade.
- A score of 28 or more points means a restaurant receives a C grade.

A color coded histogram of our data should reflect this.
- Create a histogram with 25 bins.
- Color the border of your rectangles white so that you can see where one starts and the next begins.
- Color the inside of the rectangles with the colors of the corresponding letter grade sign.
- Apply themes and labels to make your plot aesthetically pleasing.

````r
ggplot(restaurants, aes(x=SCORE, fill=GRADE)) + 
    geom_histogram(bins=25, color='white') + 
    scale_fill_manual(
        name= "Inspection Grade",
        values=c('royalblue','green3','orange')
    ) +
    ggtitle('Distribution of Inspection Scores') +
    xlab('Inspection Score') +
    ylab('Number of Restaurants') +
    my_themes
````
