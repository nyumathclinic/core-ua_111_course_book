### Exercise 1: Vectorized max and min

The functions `pmin()` and `pmax()` are vectorized versions of `min()` and `max()` respectively. They return a vector of values which are generated by taking an entry-by-entry minimum or maximum. For example, `pmin(c(-1,0,1,2), c(4,2,0,-2))` outputs `c(-1,0,0,-2)` and `pmax(c(-1,0,1,2), c(4,2,0,-2))` outputs `c(4,2,1,2)`. 

These functions can also be used to compare the entries of a vector to a particular number. For example, `pmin(1, c(4,2,0,-2))` outputs  `c(1,1,0,-2)` and `pmax(1, c(4,2,0,-2))` outputs `c(4,2,1,1)`.

```pmin(NUMBER_OR_VECTOR, NUMBER_OR_VECTOR)```

```pmax(NUMBER_OR_VECTOR, NUMBER_OR_VECTOR)```
